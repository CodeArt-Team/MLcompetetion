```python
def character_col_extraction(df_smile):
    import pandas as pd

    def total_count_character_make_column(smiles_str=""):
        """

        """
        if smiles_str:
            smiles = smiles_str

        char_counts = {}
        for char in smiles:
            if char in char_counts:
                char_counts[char] += 1
            else:
                char_counts[char] = 1

        df = pd.DataFrame.from_dict(char_counts, orient='index', columns=['Count'])
        df.index.name = 'Character'
        return df.T

    all_char_counts = pd.DataFrame()

    for i in df_smile['Smiles']:
        char_counts = total_count_character_make_column(smiles_str=i)
        all_char_counts = all_char_counts.add(char_counts, fill_value=0)

    char_column = all_char_counts.columns.tolist()
    char_column.append("Smiles")

    ####-------------------

    def count_character(smiles_str=""):
        """
        """
        if smiles_str:
            smiles = smiles_str
        char_counts = {'left_bracket': 0, 'right_bracket': 0} # Ï¥àÍ∏∞Ìôî
        for char in smiles:
            if char in char_counts:
                if char == "[":
                    print("[ÏûÖÎãàÎã§")
                    char_counts["left_bracket"] += 1
                elif char =="]":
                    char_counts["right_bracket"] += 1
                else:
                    char_counts[char] += 1
            else:
                
                if char == "[":
                    char_counts["left_bracket"] += 1
                elif char =="]":
                    char_counts["right_bracket"] += 1
                else:
                    char_counts[char] = 1
        return char_counts
    # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±: Í∞Å Î¨∏ÏûêÏó¥ÏùÑ Ïó¥Î°ú Í∞ÄÏßÄÎäî Îπà Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ

    all_char_counts = pd.DataFrame(columns=char_column)

    # SMILES Î¨∏ÏûêÏó¥ ÏàúÌöå
    for i, smiles in enumerate(df_smile['Smiles']):
        char_counts = count_character(smiles_str=smiles)
        row = pd.Series({k: v for k, v in char_counts.items()})  # ÎîïÏÖîÎÑàÎ¶¨Î•º ÏãúÎ¶¨Ï¶àÎ°ú Î≥ÄÌôò
        row = row.reindex(char_column, fill_value=0)  # Ïª¨Îüº ÏàúÏÑú ÎßûÏ∂îÍ≥† ÏóÜÎäî Í∞í 0ÏúºÎ°ú Ï±ÑÏö∞Í∏∞
        row['Smiles'] = smiles  # SMILES Î¨∏ÏûêÏó¥ Ï∂îÍ∞Ä
        all_char_counts.loc[i] = row  # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Ìñâ Ï∂îÍ∞Ä

    df_display_centered(all_char_counts.head())  ## Cl ,cnn,ccc Îì±Îì± Î∂ÑÎ¶¨ÏïàÎêòÎäî Î¨∏Ï†úÏûàÏùå. 
    return all_char_counts

all_char_counts = character_col_extraction(df_smile)
```
