```python
test = test
import joblib

# ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
loaded_model = joblib.load("XG_model")  # ëª¨ë¸ íŒŒì¼ ì´ë¦„

# ëª¨ë¸ ì‚¬ìš©
char_column=['#', '(', ')', '+', '-', '.', '/', '1', '2', '3', '4', '5', '6', '7', '8', '=', '@', 'B', 'C', 'F', 'H', 'N', 'O', 'P', 'S', 'left_bracket', '\\', 'right_bracket','c', 'l', 'n', 'o', 'r', 's']
# loaded_model.predict()
test_col_change = pd.DataFrame(columns=char_column)
def piC50_to_IC50(pIC50):
    import numpy as np
    pIC50=abs(pIC50)
    IC50=10**(-pIC50)*(10**9)
    # print(pIC50)
    return IC50

def count_character(smiles_str=""):
    """
    """
    if smiles_str:
        smiles = smiles_str
    char_counts = {'left_bracket': 0, 'right_bracket': 0} # ì´ˆê¸°í™”
    for char in smiles:
        if char in char_counts:
            if char == "[":
                print("[ì…ë‹ˆë‹¤")
                char_counts["left_bracket"] += 1
            elif char =="]":
                char_counts["right_bracket"] += 1
            else:
                char_counts[char] += 1
        else:
            
            if char == "[":
                char_counts["left_bracket"] += 1
            elif char =="]":
                char_counts["right_bracket"] += 1
            else:
                char_counts[char] = 1
    return char_counts


# SMILES ë¬¸ìì—´ ìˆœíšŒ
for i, smiles in enumerate(test['Smiles']):
    char_counts = count_character(smiles_str=smiles)
    row = pd.Series({k: v for k, v in char_counts.items()})  # ë”•ì…”ë„ˆë¦¬ë¥¼ ì‹œë¦¬ì¦ˆë¡œ ë³€í™˜
    row = row.reindex(char_column, fill_value=0)  # ì»¬ëŸ¼ ìˆœì„œ ë§ì¶”ê³  ì—†ëŠ” ê°’ 0ìœ¼ë¡œ ì±„ìš°ê¸°
    row['Smiles'] = smiles  # SMILES ë¬¸ìì—´ ì¶”ê°€
    test_col_change.loc[i] = row  # ë°ì´í„°í”„ë ˆì„ì— í–‰ ì¶”ê°€

# df_display_centered(test_col_change)

# # ì»¬ëŸ¼ ì´ë¦„ ë³€ê²½
print(test_col_change.iloc[0,:].to_numpy())
```
