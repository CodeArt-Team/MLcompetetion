```python
import pandas as pd
df_smile = pd.DataFrame(input_df['Smiles'])
df_display_centered(df_smile.head())
```

```python
import pandas as pd

def total_count_character_make_column(smiles_str=""):
    """

    """
    if smiles_str:
        smiles = smiles_str

    char_counts = {}
    for char in smiles:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1

    df = pd.DataFrame.from_dict(char_counts, orient='index', columns=['Count'])
    df.index.name = 'Character'
    return df.T

all_char_counts = pd.DataFrame()

for i in df_smile['Smiles']:
    char_counts = total_count_character_make_column(smiles_str=i)
    all_char_counts = all_char_counts.add(char_counts, fill_value=0)

char_column = all_char_counts.columns.tolist()
char_column.append("Smiles")

####-------------------

def count_character(smiles_str=""):
    """
    """
    if smiles_str:
        smiles = smiles_str
    char_counts = {}
    for char in smiles:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1
    return char_counts
# ë°ì´í„°í”„ë ˆì„ ìƒì„±: ê° ë¬¸ìì—´ì„ ì—´ë¡œ ê°€ì§€ëŠ” ë¹ˆ ë°ì´í„°í”„ë ˆì„

all_char_counts = pd.DataFrame(columns=char_column)

# SMILES ë¬¸ìì—´ ìˆœíšŒ
for i, smiles in enumerate(df_smile['Smiles']):
    char_counts = count_character(smiles_str=smiles)
    row = pd.Series({k: v for k, v in char_counts.items()})  # ë”•ì…”ë„ˆë¦¬ë¥¼ ì‹œë¦¬ì¦ˆë¡œ ë³€í™˜
    row = row.reindex(char_column, fill_value=0)  # ì»¬ëŸ¼ ìˆœì„œ ë§ì¶”ê³  ì—†ëŠ” ê°’ 0ìœ¼ë¡œ ì±„ìš°ê¸°
    row['Smiles'] = smiles  # SMILES ë¬¸ìì—´ ì¶”ê°€
    all_char_counts.loc[i] = row  # ë°ì´í„°í”„ë ˆì„ì— í–‰ ì¶”ê°€

df_display_centered(all_char_counts.head())  ## Cl ,cnn,ccc ë“±ë“± ë¶„ë¦¬ì•ˆë˜ëŠ” ë¬¸ì œìˆìŒ. 

```
