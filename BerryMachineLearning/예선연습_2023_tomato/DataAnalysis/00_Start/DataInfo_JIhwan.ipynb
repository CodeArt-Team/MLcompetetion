```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def check_correlation(dataframe, method='pearson', figsize=(20, 16)):
    """
    Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Î™®Îì† Ïª¨Îüº Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Í≥ÑÏÇ∞ÌïòÍ≥† ÌûàÌä∏ÎßµÏúºÎ°ú ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
    
    Parameters:
    dataframe (pd.DataFrame): ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÌôïÏù∏Ìï† Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
    method (str): ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Í≥ÑÏÇ∞ Î∞©Î≤ï ('pearson', 'spearman', 'kendall')
    figsize (tuple): Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞
    
    Returns:
    pd.DataFrame: ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Îß§Ìä∏Î¶≠Ïä§
    """
    # ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Í≥ÑÏÇ∞
    corr_matrix = dataframe.corr(method=method)
    
    # ÌûàÌä∏Îßµ Í∑∏Î¶¨Í∏∞
    plt.figure(figsize=figsize)
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f')
    plt.title(f'Correlation Heatmap ({method})')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    return corr_matrix

# ÏûÖÎ†• Î≥ÄÏàòÏôÄ Ï∂úÎ†• Î≥ÄÏàòÎ•º Î™®Îëê Ìè¨Ìï®Ìïú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
all_cols = input_cols + output_cols
correlation_data = input_data[all_cols]

# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ ÏÉùÏÑ±
correlation_matrix = check_correlation(correlation_data)

# ÌäπÏ†ï Î≥ÄÏàòÎì§Í≥ºÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎßå ÌôïÏù∏ÌïòÍ≥† Ïã∂Îã§Î©¥:
# Ïòà: Ï∂úÎ†• Î≥ÄÏàòÏôÄ Îã§Î•∏ Î™®Îì† Î≥ÄÏàòÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ
output_correlation = correlation_matrix[output_cols].sort_values(by=output_cols[0], ascending=False)
print("Correlation with output variables:")
print(output_correlation)
```
